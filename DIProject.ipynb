```python
from sklearn import datasets, linear_model
import csv
import pandas as pd
import numpy as np
import datetime as dt
from matplotlib import pyplot
from pandas import Series
```

```python
weather_df = pd.read_csv('weather.csv')
elect_df = pd.read_csv('ny.electricity.csv')
temp_df = weather_df[['DATE','TAVG','TMAX','TMIN']]
```

```python
temp_df.index = pd.to_datetime(weather_df['DATE'])
```

```python
elect_df['date'] = elect_df.apply(lambda row: str(row.YEAR)+'/' + str(row.MONTH), axis=1)
```

```python
elect_df = elect_df[['date', 'GENERATION (Megawatthours)']]
elect_df.index = pd.to_datetime(elect_df['date'])
```

```python
elect_df['GENERATION (Megawatthours)'] = elect_df['GENERATION (Megawatthours)'].apply(lambda x: int(x.replace(',','')))
```

```python
electricity = Series(data = elect_df['GENERATION (Megawatthours)'].values)
electricity.plot()
```
